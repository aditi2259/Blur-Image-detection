```python
import cv2
import numpy as np
import os
from google.colab.patches import cv2_imshow

def is_blurry(image_path, threshold=100):

    image = cv2.imread("/content/blur.jpeg")

    if image is None:
        print("Error: Unable to load image.")
        return None

    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    laplacian_var = cv2.Laplacian(gray, cv2.CV_64F).var()

    # If variance is lower than threshold, the image is blurry
    if laplacian_var < threshold:
        return True, laplacian_var
    else:
        return False, laplacian_var

image_path = "/content/blur.jpeg"

result = is_blurry(image_path)

if result is not None:
    blurry, variance = result
    if blurry:
        print(f"The image is blurry. Variance: {variance:.2f}")
    else:
        print(f"The image is clear. Variance: {variance:.2f}")
cv2_imshow(image)
```
